**Setting Up Discriminator's & Generator's Dictionary Parameters**

```python
"""
Parameters Disc:
dis_in_dim (int) : dimension of date fed into discriminator
dis_out_dim (int) : dimension of discriminator's output
dis_hidden_layers (list) : list of neurons for each hidden layer
leaky_relu (float) : lealy relu to be used for each hidden layer
dropout (float) : dropout value to be used for each hidden layer"""
from torch import nn,cuda
loss=nn.BCELoss()
lr=0.0002 
is_img_data=True
epochs=50
batch_size=100

dis_in_dim=28*28 # mnist features 28 X 28
dis_gen_dict={"dis_in_dim":dis_in_dim,
          "dis_out_dim":1, 
          "dis_hidden_layers":[1024,512,256],
          "dis_leaky_relu":0.2, 
          "dis_dropout":0.3,
         
          "gen_in_dim":100,
          "gen_out_dim":dis_in_dim,
          "gen_hidden_layers":[256,512,1024],
          "gen_leaky_relu":0.2,
         
          "loss":loss}

discriminator=Discriminator(**dis_gen_dict).cuda()
optim_dis=optim.Adam(discriminator.dis_layers.parameters(), lr=lr)


"""    
Parameters Gen:
gen_in_dim (int) : dimension of noise fed into generator
gen_out_dim (int) : dimension of generator's output
gen_hidden_layers (list) : list of neurons for each hidden layer
leaky_relu (float) : lealy relu to be used for each layers
"""

generator=Generator(**dis_gen_dict).cuda()

optim_gen=optim.Adam(generator.gen_layers.parameters(), lr=lr)
dis_gen_dict["optim_dis"]=optim_dis
dis_gen_dict["optim_gen"]=optim_gen
dis_gen_dict["dis_obj"]=discriminator
dis_gen_dict["gen_obj"]=generator
print("Done")
```

```python
modeler=Modeler(epochs,batch_size,is_img_data)
modeler.train(**dis_gen_dict)
```
